````python
# Function for ploting the decision boundaries of a model
def plot_decision_regions(X, y, classifier, resolution=0.01, title=""):

    # setup marker generator and color map
    markers = ('.', '.')
    colors = ['blue', 'red']
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.title(title)
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
    plt.show()
````

````python
def visualize_data(X, y, title_prefix=""):
    if title_prefix is None or title_prefix == "":
        title_prefix = ""
    else:
        title_prefix = f"{title_prefix} - "

    # prepare data
    classes = np.unique(y_training)
    X_1000 = X[:1000, :]
    y_1000 = y[:1000]

    # hist for first feature - first 1000 data points:
    plt.figure(figsize=(8, 5))

    for i in range(len(classes)):
        data = X_1000[y_1000 == classes[i]][:, 0]
        plt.hist(data, bins=20, alpha=0.5, label=['class={}'.format(i)])  # `density=False` would make counts

    plt.xlabel('x1 feature')
    plt.ylabel('Amount')
    plt.title(f'{title_prefix}histogram of first 1000 data points of each class')
    plt.grid()
    plt.legend()
    plt.show()

    # hist for first feature - all data points:
    plt.figure(figsize=(8, 5))

    for i in range(len(classes)):
        data = X[y == classes[i]][:, 0]
        plt.hist(data, bins=20, alpha=0.5, label=['class={}'.format(i)])  # `density=False` would make counts

    plt.xlabel('x1 feature')
    plt.ylabel('Amount')
    plt.title(f'{title_prefix}histogram of all data points of each class')
    plt.grid()
    plt.legend()
    plt.show()

    # scatter for both features - first 1000 data points:
    plt.figure(figsize=(8, 5))
    for i in range(len(classes)):
        data = X_1000[y_1000 == classes[i]]
        plt.scatter(data[:, 0], data[:, 1], label='class={}'.format(i))

    plt.title(f'{title_prefix}scatter plot for first 1000 data points of each class')
    plt.legend()
    plt.xlabel('x1 feature units')
    plt.ylabel('x2 feature units')
    plt.grid()
    plt.show()

    # scatter for both features - all data points:
    plt.figure(figsize=(8, 5))
    for i in range(len(classes)):
        data = X[y == classes[i]]
        plt.scatter(data[:, 0], data[:, 1], label='class={}'.format(i))

    plt.title(f'{title_prefix}scatter plot for all data points of each class')
    plt.legend()
    plt.xlabel('x1 feature units')
    plt.ylabel('x2 feature units')
    plt.grid()
    plt.legend()
    plt.show()
````

## Cross Validation (10 points)

Use 5-fold cross validation in order to find the best eps and eta params from the given lists.

Open `hw4.py` and implement the function `cross_validation`.

Note the following:
1. This function should not change the original data.
1. You should shuffle the data before splitting in order to get correct results.

````python
from hw4 import cross_validation
````

````python
#### Your code here ####
etas = [0.05, 0.005, 0.0005, 0.00005, 0.000005]
epss = [0.01, 0.001, 0.0001, 0.00001, 0.000001]
folds = 5

random_state = 1
acc_dict = {}
for eta in etas:
    for eps in epss:
        lor = LogisticRegressionGD(eta=eta, eps=eps, random_state=random_state)
        acc = cross_validation(X_training, y_training, folds, lor, random_state=random_state)
        acc_dict[(eta, eps)] = acc
        print(f"Accuracy with eta={eta}, eps={eps}:  {acc:.6f}")

best_params = max(acc_dict, key=acc_dict.get)
best_eta = best_params[0]
best_eps = best_params[1]
````

````python
print(f"Best learning rate (eta): {best_eta}")
print(f"Best epsilon: {best_eps}")

best_acc = acc_dict[best_params]
print(f"Accuracy with the best parameters: {best_acc:.4f}")
````

````python
## your tests here ##
def visualize_datasets(datasets):
    # Unpack the datasets
    dataset_a_features = datasets['dataset_a_features']
    dataset_a_labels = datasets['dataset_a_labels']
    dataset_b_features = datasets['dataset_b_features']
    dataset_b_labels = datasets['dataset_b_labels']

    # Define colors for labels
    colors_a = ['red' if label == 0 else 'blue' for label in dataset_a_labels]
    colors_b = ['red' if label == 0 else 'blue' for label in dataset_b_labels]

    # 2D Visualization
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))

    axes[0, 0].scatter(dataset_a_features[:, 0], dataset_a_features[:, 1], c=colors_a)
    axes[0, 0].set_title('Dataset A: Feature 1 vs Feature 2')
    axes[0, 1].scatter(dataset_a_features[:, 0], dataset_a_features[:, 2], c=colors_a)
    axes[0, 1].set_title('Dataset A: Feature 1 vs Feature 3')
    axes[0, 2].scatter(dataset_a_features[:, 1], dataset_a_features[:, 2], c=colors_a)
    axes[0, 2].set_title('Dataset A: Feature 2 vs Feature 3')

    axes[1, 0].scatter(dataset_b_features[:, 0], dataset_b_features[:, 1], c=colors_b)
    axes[1, 0].set_title('Dataset B: Feature 1 vs Feature 2')
    axes[1, 1].scatter(dataset_b_features[:, 0], dataset_b_features[:, 2], c=colors_b)
    axes[1, 1].set_title('Dataset B: Feature 1 vs Feature 3')
    axes[1, 2].scatter(dataset_b_features[:, 1], dataset_b_features[:, 2], c=colors_b)
    axes[1, 2].set_title('Dataset B: Feature 2 vs Feature 3')

    plt.show()

    # 3D Visualization
    fig = plt.figure(figsize=(12, 6))

    ax = fig.add_subplot(121, projection='3d')
    ax.scatter(dataset_a_features[:, 0], dataset_a_features[:, 1], dataset_a_features[:, 2], c=colors_a)
    ax.set_title('Dataset A: 3D Visualization')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_zlabel('Feature 3')

    ax = fig.add_subplot(122, projection='3d')
    ax.scatter(dataset_b_features[:, 0], dataset_b_features[:, 1], dataset_b_features[:, 2], c=colors_b)
    ax.set_title('Dataset B: 3D Visualization')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_zlabel('Feature 3')

    plt.show()

# Generate the datasets
datasets = generate_datasets()

# Visualize the datasets
visualize_datasets(datasets)
````
